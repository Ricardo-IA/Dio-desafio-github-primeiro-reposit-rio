# **Copa do Mundo de Futebol FIFA 2022**

```python
#importando a biblioteca Pandas

import pandas as pd
import matplotlib.pyplot as plt

```

Carregando arquivo csv Copa_22, baixado do https://www.kaggle.com/datasets/piterfm/fifa-football-world-cup?select=matches_2022.csv

```python
df = pd.read_csv('/content/Copa_2022.csv')
```

```python
#visualizando as 5 primeiras colunas
df.head()
```

```python
#Renomeando colunas do dataframe
df = df.rename(columns={"home_team":"Time da casa","away_team":"Time Visitante","home_score":"Pontuação em Casa","home_xg":"Casa XG","home_penalty":"Penalidade em casa","away_score":"Resultado fora de casa","away_xg":"Fora XG","away_penalty":"Pênalti fora de casa","home_manager":"Técnico","home_captain":"Capitão do time","home_goals":"Gols em casa", "golos_fora":"Gols fora de casa","Attendance":"Comparecimento","Venue":"Local","Officials":"Funcionarios","Date":"Data","Score":"Pontuação","Referee":"Juiz","Notes":"Notas","Host":"Hospedeiro","Encontro":"Encontro","Year":"Ano" })
df.head(26)
```

```python
#informações estatistica
df.describe()
```

```python
#convertendo a notação cientifica
pd.options.display.float_format ='{:20,.2f}'.format
df.head(26)
```

```python
#selecionando colunas do dataframe Pais, Técnico e Capitão
df = df.loc[:,["home_team","home_manager","home_captain"]]
df.head(26)
```

```python
#total de torcida no estádio
df.groupby("home_team")["Attendance"].sum().sort_values(ascending = False)
```

```python
from pickle import TRUE
#Grafico do total de torcedores no estádio
df.groupby("home_team")["Attendance"].sum().sort_values(ascending = True).plot.bar(title ="total de torcedores no estádio")
plt.xlabel ("Seleções")
plt.ylabel("Quantidade de torcedores");
plt.grid()
```
